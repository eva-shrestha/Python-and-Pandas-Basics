```python
# 5. Find the maximum value in the Sugars column for each manufacturer.
```

```python
group = cereals_df.groupby(['Manufacturer'])
max_sugar = group['Sugars'].max()
max_sugar
```

```python
# 6. Find the minimum value in the Fiber column for each manufacturer.
```

```python
min_fiber = group['Fiber'].min()
min_fiber
```

```python
# 7. Extract the cereal with the lowest amount of grams of sugar per manufacturer in a newDataFrame.
```

```python
low_sugar = cereals_df.groupby(['Manufacturer']).min(['Sugars'])
low_sugar
```

```python
new_df = pd.DataFrame(low_sugar)
new_df
```

```python
# Each address consists of a street,a city, a state, and a zip code. 
# Your challenge is to separate these four values; 
# assign them to new Street, City, State, and Zip columns; and then remove the Address column.
```

```python
# Split the Address column into four separate columns
seperate_address = customers_df['Address'].str.split(',', expand=True)
seperate_address.columns = ['Street', 'City', 'State', 'Zip']
```

```python
# Add the new columns to the DataFrame using concat
customers_df = pd.concat([customers_df, seperate_address], axis=1)
```

```python
customers_df.drop('Address', axis=1, inplace=True)
```
