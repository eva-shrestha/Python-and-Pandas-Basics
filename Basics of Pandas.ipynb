# Series

```python
a = [1,2,4,5]
```

```python
myvar = pd.Series(a)
```

```python
print(myvar)
```

```python
#using labels to print
print(myvar[3])
```

```python
#we can also create own labels
mylabel = pd.Series(a, index=["x","y","z","b"])
```

```python
print(mylabel["y"])
```

```python
#key value objects as series
calories = {"day1": 420, "day2": 380, "day3": 390}
var = pd.Series(calories)
```

```python
print(var)
```

```python
#Series is a one-dimensional labeled array capable of holding any data type
#series is better than list because it is faster and smaller memory wise
#A Series also offers method to manipulate and describe data which a list has not.
```

# Find and drop duplicates from the titanic dataset

```python
data = pd.read_csv('titanic_data.csv')
```

```python
#finding duplicates
duplicate = data[data.duplicated()]
print("Duplicate Rows")
duplicate
```

```python
#since there are no duplicate data in the titanic dataset, let us first add a duplicated row into the data and handle it.

df_dup = data.copy()

#duplicating the first row and append it to the data

dup_row = df_dup.iloc[:1]

df_dup = df_dup.append(dup_row, ignore_index = True)
df_dup
```

```python
#finding duplicate rows

df_dup[df_dup.duplicated()]
```

```python
#removing duplicate rows

df_dup.drop_duplicates()
```

#  Pandas `apply()` function.


```python
df = pd.DataFrame({'A':[1,2], 'B':[10,20]})

def square(x):
    return x*x

df1 = df.apply(square)
df1
```

```python
# apply() with lambda

df2 = df.apply(lambda x : x*10)
df2
```

# Using Filer() in data frame

```python
student = {
    'Name' : ['Ar','Sanj','Den','Bij'],
    'Courses' : ['Data', 'QA', 'UI', 'Frontend'],
    'Grade' : [100,100,90,95]
}

df = pd.DataFrame(student)
df
```

```python
# filter columns
df1 = df.filter(items=['Name','Courses'])
df1
```

```python
# filter columns using like

df2 = df.filter(like='rse',axis=1)
df2
```

```python
#filter columns using regex

df3 = df.filter(regex='e$', axis=1)
df3
```

```python
# filter() rows by index

df4 = df.filter(items = [1,3],axis=0)
df4
```

```python
#filter() rows usign like

df5 = df.filter(like='2', axis=0)
df5
```
