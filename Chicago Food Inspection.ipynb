```python
df['Name'] = df['Name'].str.strip()
```

# Change to lower case when risk is risk 1

```python
df['Name'] = df.apply(lambda x: x['Name'].lower() if x['Risk'] == 'Risk 1 (High)' else x['Name'], axis=1)
```
